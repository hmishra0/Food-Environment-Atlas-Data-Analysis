```python
X= df2[df2.columns.difference(['FIPS', 'State','County'])]
N = X.shape[0]
```

```python
from sklearn.cluster import KMeans
import sklearn.cluster as cluster
import sklearn.metrics as metrics
kmeans = KMeans(n_clusters=10, random_state=0).fit(X)
```

```python
# Determine the number of clusters
nClusters = np.zeros(10)
Elbow = np.zeros(10)
Silhouette = np.zeros(10)
TotalWCSS = np.zeros(10)
Inertia = np.zeros(10)
```

```python
for c in range(10):
    KClusters = c + 1
    nClusters[c] = KClusters
    
    kmeans = cluster.KMeans(n_clusters=KClusters, random_state=0).fit(X)

   # The Inertia value is the within cluster sum of squares deviation from the centroid
    Inertia[c] = kmeans.inertia_
    
    if (1 < KClusters & KClusters < 10):
        Silhouette[c] = metrics.silhouette_score(X, kmeans.labels_)
    else:
        Silhouette[c] = np.NaN
    
    WCSS = np.zeros(KClusters)
    nC = np.zeros(KClusters)
    
    for i in range(N):
        k = kmeans.labels_[i]
        nC[k] += 1
        diff = X.iloc[i,]  - kmeans.cluster_centers_[k]
        WCSS[k] += diff.dot(diff)
    
    Elbow[c] = 0
    for k in range(KClusters):
        Elbow[c] += WCSS[k] / nC[k]
        TotalWCSS[c] += WCSS[k]


print("N Clusters\t Inertia \t\tTotal WCSS \t\t Elbow Value \t\t Silhouette Value")
for c in range(10):
    print('{:.0f} \t\t {:.4f} \t\t {:.3f} \t\t {:.4f} \t\t {:.4f}'
          .format(nClusters[c], Inertia[c], TotalWCSS[c], Elbow[c], Silhouette[c]))
```

We can see that the largest silhouette is for cluser = 2

```python
from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

We can see that the number of clusters are two from above plot and metrics

```python
# Interpret 2 cluster solution
model = KMeans(n_clusters = 3)
model.fit(X) 
clusters = model.predict(X)
clusters
```

```python
# Visulaizing clusters
from sklearn.decomposition import PCA 
pca_3 = PCA(3)
plot_columns = pca_3.fit_transform(X)
plt.scatter(x=plot_columns[:,0], y=plot_columns[:,1], c=model.labels_,)
plt.xlabel('Canonical variable 1')
plt.ylabel('Canonical variable 2')
plt.title('Scatterplot of Canonical Variables for 3 Clusters')
plt.show()
```

```python
X.reset_index(level=0, inplace=True)
```

```python
cluslist=list(X['index'])
```

```python
labels=list(model.labels_)
# combine index variable list with cluster assignment list into a dictionary
newlist=dict(zip(cluslist, labels))
# convert newlist dictionary to a dataframe
newclus=pd.DataFrame.from_dict(newlist, orient='index')
newclus
# rename the cluster assignment column
newclus.columns = ['cluster']
newclus.reset_index(level=0, inplace=True)

# by the index variable
merged_df = pd.merge(X, newclus, on='index')
merged_df.head(n=100)
# cluster frequencies
merged_df.cluster.value_counts()
```

```python
clustergrp = merged_df.groupby('cluster').mean()
print ("Clustering variable means by cluster")
pd.set_option('display.max_columns', None)
clustergrp
```
